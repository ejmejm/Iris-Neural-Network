```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn import datasets
```

```python
np.random.seed(0)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoidPrime(x):
    return sigmoid(x) * (1-sigmoid(x))

t = np.arange(-5., 5., 0.5)
plt.plot(t, sigmoid(t))
plt.plot(t, sigmoidPrime(t))
plt.show()
```

```python
iris = datasets.load_iris()
irisX = np.asarray(iris.data[:, :]).reshape(150, 4)
pre_irisY = np.asarray(iris.target).reshape(150)
irisY = np.zeros((150, 3))

for i in range(len(pre_irisY)):
    irisY[i][pre_irisY[i]] = 1
```

```python
plt.plot(irisX)
plt.plot(irisY)
plt.show()
plt.scatter()
```

```python
permutation = np.random.permutation(irisX.shape[0])
irisX = irisX[permutation]
irisY = irisY[permutation]
```

```python
plt.plot(irisX)
plt.plot(irisY)
plt.show()
```

```python
nIn = 4
nHidden = 10
nOut = 3
alpha = 0.1
nEpochs = 500
```

```python
def nn_forward(X, W1, W2):
    x = X.reshape(-1, nIn)
    
    layer2 = np.dot(x, W1)
    layer2 = sigmoid(layer2)
    
    out = np.dot(layer2, W2)
    out = sigmoid(out)
    
    return out
```

```python
def train(X, Y, W1, W2):
    x = X.reshape(-1, nIn)
    Y = Y.reshape(-1, nOut)
    
    layer2 = np.dot(x, W1)
    a_layer2 = sigmoid(layer2)
    
    out = np.dot(a_layer2, W2)
    a_out = sigmoid(out)
    
    loss = np.sum(0.5 * (Y - a_out) ** 2)
    
    delta_loss = (Y - a_out) * -sigmoidPrime(out)
    delta_W2 = np.dot(layer2.T, delta_loss)
    
    delta_loss = np.dot(delta_loss, W2.T) * sigmoidPrime(layer2)
    delta_W1 = np.dot(X.T.reshape(nIn, 1), delta_loss)
    
    W2 = W2 - alpha * delta_W2
    W1 = W1 - alpha * delta_W1
    
    return W1, W2, loss
```

```python
losses = []
W1 = np.random.rand(nIn, nHidden) * 0.5
W2 = np.random.rand(nHidden, nOut) * 0.5
```

```python
trainX = irisX[:-15]
testX = irisX[-15:]

trainY = irisY[:-15]
testY = irisY[-15:]
```

```python
for j in range(nEpochs):
    for i in range(trainX.shape[0]):
        W1, W2, loss = train(trainX[i], trainY[i], W1, W2)
    if j % 50 == 0:
        print("Epoch {} Loss: {}".format(j, loss))
    if j % 5 == 0:
        losses.append(loss)
            
plt.plot(losses)
plt.show()
```

```python
correct = 0
total = len(testX)

for i in range(total):
    pred = np.argmax(nn_forward(testX[i], W1, W2))+1
    actual = np.argmax(testY[i])+1
    
#     print("Prediction: Type {}".format(pred))
#     print("Actual: Type {}\n".format(actual))
    
    if pred == actual:
        correct +=1
        
print("Accuracy: {}%".format(correct/total * 100))
```
